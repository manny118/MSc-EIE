```python
# Author: Emmanuel Akinrintoyo
# import the Python libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function for plotting the figures
def plotFigure(fig, title, xlabel, ylabel, font_size):
    plt.plot(fig[0], fig[1], color='b')
    plt.title(title, fontsize=font_size, fontname="Arial")
    plt.xlabel(xlabel, fontsize=font_size, fontname="Arial")
    plt.ylabel(ylabel, fontsize=font_size, fontname="Arial")
    title = title.replace(" ", "_") + str(font_size)
    
```

```python
wavelength = np.array([1431, 1451, 1471, 1491, 1511, 1531, 1551, 1571, 1591,
                       1611])
att = np.array([0.263, 0.25, 0.238, 0.229, 0.221, 0.215, 0.211, 0.208, 0.208,
                0.208])
x = wavelength
y = att

fit = np.polyfit(wavelength, att, 2)
a, b, c = fit[0], fit[1], fit[2]
fit_equation = a * np.square(x) + b * x + c
```

```python
plt.plot(x, y, 'o', color='red')
plt.plot(x, fit_equation, color='b', alpha=0.4)

font_size = 16
plt.ylabel("Attenuation (dB/km)", fontsize=font_size, fontname="Arial")
plt.xlabel("Wavelength (nm)", fontsize=font_size, fontname="Arial")
title = title.replace(" ", "_") + str(font_size)
```
